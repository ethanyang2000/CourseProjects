```python
import pandas as pd

df = pd.read_csv('popular_board_game.csv')
res = {}
ans = {}
for index, row in df.iterrows():
    if row['category'] not in res.keys():
        res[row['category']] = {'solo':{'play':[], 'rating':[], 'num':[], 'name':[]}, 'unsolo':{'play':[], 'rating':[], 'num':[], 'name':[]}}
    if row['min_players'] == 1:
        res[row['category']]['solo']['name'].append(row['name'])
        res[row['category']]['solo']['rating'].append(row['avg_rating'])
        res[row['category']]['solo']['num'].append(row['num_ratings'])
        res[row['category']]['solo']['play'].append(row['playtime_num'])
    else:
        res[row['category']]['unsolo']['name'].append(row['name'])
        res[row['category']]['unsolo']['rating'].append(row['avg_rating'])
        res[row['category']]['unsolo']['num'].append(row['num_ratings'])
        res[row['category']]['unsolo']['play'].append(row['playtime_num'])

data = {}
        
for key, value in res.items():
    print(key)
    for k, v in value.items():      
        data[(k, key)] = [len(v['rating'])]
        if len(v['num']) == 0:
            print("max value: NA")
            data[(k, key)].append("NA")
        else:
            print("max idx: ", v['num'].index(max(v['num'])))
            print("max name: ", v['name'][v['num'].index(max(v['num']))])
            data[(k, key)].append(v['name'][v['num'].index(max(v['num']))])
        if len(v['rating']) == 0:
            print("avg_rating: ", 0)
            data[(k, key)].append('NA')
        else:
            print("avg_rating: ", sum(v['rating'])/len(v['rating']))
            data[(k, key)].append("{:.2f}".format(sum(v['rating'])/len(v['rating'])))
        if len(v['play']) == 0:
            print("avg_play: ", 0)
            data[(k, key)].append('NA')
        else:
            print("avg_play: ", sum(v['play'])/len(v['play']))
            data[(k, key)].append("{:.2f}".format(sum(v['play'])/len(v['play'])))
ans = pd.DataFrame(data, index=['Total Num of Games', 'Game with Highest Num of Ratings', 'Avg Rating', 'Avg Playtime'])
print(ans)
```
